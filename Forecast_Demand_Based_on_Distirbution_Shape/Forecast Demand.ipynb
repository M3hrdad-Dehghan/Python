### **Import libraries**

```python
import numpy as np
import seaborn as sns
from scipy import stats
```

### **Business Fact**

Considren in a hospital, out of 5000 patients, 20% pass away. So, the binomial distribution is:

```python
dist = stats.binom(5000, 0.2)
```

Based on the distrobution, we need to create data. This data in totally randomly selected between 1 to 5000.

```python
hospital_data = dist.rvs(size = 5000)
hospital_data
```

Now, let's look at the distribution shape:

```python
sns.displot(data = hospital_data)
```

### **Business Question**

The manager is committed to equipping the hospital with essential equipment to save as many patients' lives as possible. However, in order to negotiate costs with suppliers, she would like to determine the maximum number of beds the hospital needs each day to accommodate all patients requiring this equipment to prevent fatalities.

```python
max_probability = 0
max_beds = 0

for i in range(5000):
  probability = dist.cdf(i) * 100
  if probability > max_probability:
    max_probability = probability
    max_beds = i

print(f"The maximum number of beds is {max_beds} which happens with probability {max_probability:.2f}%")
```

Providing 1241 beds allows us to cover all patients and prevent any chance of death. With a probability of 100%, the hospital will not have more than 1241 deceased patients in a day.
